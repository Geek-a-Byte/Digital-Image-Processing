```python


import cv2
import matplotlib.pyplot as plt 
import math
import numpy as np

	
# Read the image 
img = cv2.imread('images/coin.jpg', 0) 

# Obtain number of rows and columns 
# of the image 
m, n = img.shape 

# Develop laplacian filter(3, 3) mask 
mask = np.ones([3, 3], dtype = int) 
#mask = mask * 1

mask[0, 0] = 0
mask[0, 2] = 0
mask[1, 1] = -4
mask[2, 0] = 0
mask[2, 2] = 0


# Convolve the 3X3 mask over the image 
img_filtered = np.zeros([m, n]) 

for i in range(1, m-1): 
	for j in range(1, n-1): 
		temp = img[i-1, j-1]*mask[0, 0]+img[i-1, j]*mask[0, 1]+img[i-1, j + 1]*mask[0, 2]+img[i, j-1]*mask[1, 0]+ img[i, j]*mask[1, 1]+img[i, j + 1]*mask[1, 2]+img[i + 1, j-1]*mask[2, 0]+img[i + 1, j]*mask[2, 1]+img[i + 1, j + 1]*mask[2, 2] 
		
		img_filtered[i, j] = temp 
		
img_filtered = img_filtered.astype(np.uint8) 

#formula for image sharpening = f(x,y) + c[v^2f(x,y)] = image + c * filtered image
# c = -1 if center of the mask is negative, else 1
# sharpen the image by adding filtered image with original image 
img_new = np.zeros([m, n]) 
img_new = img - img_filtered



# Display the images
plt.imshow(img) 
plt.show()

# Display the images
plt.imshow(img_filtered) 
plt.show()

# Display the images
plt.imshow(img_new) 
plt.show()
```
